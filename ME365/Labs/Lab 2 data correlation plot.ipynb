```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
path = '~/Documents/ME365/Lab2/'

dat = pd.DataFrame(data = pd.read_excel(path +'Lab 2 Section 2 Final test data.xlsx'))
print dat
dat = dat.drop(labels = ['Sample #','Unnamed: 7'],axis = 1)
dat = dat.drop(labels = 9,axis = 0)
```

```python
fig = sns.pairplot(dat[1:],hue = 'Material')
fig.fig.suptitle('Exploring Correlations of Dataset',y=1)
fig.savefig('lab2_correlation.png')
```
