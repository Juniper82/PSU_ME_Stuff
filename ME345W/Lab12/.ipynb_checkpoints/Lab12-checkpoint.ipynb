```python
import pandas as pd
import numpy as np
import scipy.fftpack as fft
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Lab_Bench_3_Signal_1.csv')
```

```python
data.columns = ['freq_index','Yk']
```

```python
fig,ax = plt.subplots(figsize = [15,10])

ax.plot(data.freq_index,abs(data.Yk))
ax.set_xlabel(r'$\bf Frequency \ Index $',fontsize = 24)
ax.set_title('FFT Plot For Lab Bench 3 Signal 1',fontsize = 24,fontweight = 1000)
ax.set_ylabel(r'$\bf |Y_k|$',fontsize = 24)

#ax.set_xticks()
labels = [data.freq_index[i] for i in range(len(data)) if abs(data.Yk[i]) > 100]
labels = [labels[0],labels[1],1000,labels[2],2000,3000,labels[3],4000,labels[4],5000]
ax.set_xticks(labels)
plt.setp(ax.get_xticklabels(), fontsize=13.5)
plt.setp(ax.get_yticklabels(), fontsize=13.5)
ax.grid()
plt.show()
fig.savefig('MysterySignal.png')
```
