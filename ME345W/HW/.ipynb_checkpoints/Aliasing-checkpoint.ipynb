```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
plt.rcParams['axes.grid'] = True # Turns grid lines on globally
%matplotlib inline
```

```python
def getComponents(y,fs,t):
    Y = np.round(np.fft.fft(y),5)
    deltaf = fs/len(t)
    for i in range(len(Y)):
        if Y[i].real != 0 or Y[i].imag != 0:
            print(i*deltaf,Y[i].real,str(Y[i].imag)+'i')
    return Y,deltaf
```

A certain “music” signal is comprised of five “tones”:

v(t)=0.5sin(10000πt) + 0.4sin(20000πt) + 0.2sin(36000πt) + 0.3sin(48000πt) +
0.4sin(50000πt)

(Recall the form of a sinewave signal as: f(t) = A sin(2πft).)
Answer the following:

a) What frequencies are present in the original signal?

b) If v(t) is sampled at 44.1kHz, what frequencies will be present in the sampled
version of the signal?

c) What is the minimum frequency at which v(t) could be sampled to avoid
aliasing?

```python
f1 = 10e3/2
f2 = 20e3/2
f3 = 36e3/2
f4 = 48e3/2
f5 = 50e3/2
print('a) The following 5 frequencies are present in the original signal: \n'  
      'f1 =',f1,'\nf2 =',f2,'\nf3 =',f3,'\nf4 =',f4,'\nf5 =',f5)
```

```python
fs = 44.1e3
fN = fs/2
print('b) Nyquist frequency, fN = ',fN,'\nAny frequency higher than fN will result in an alias being present. \n' 
      'f1,f1,f3 will be present along with two aliases')
print('c) The minimum sampling frequency to avoid aliasing should be 50.001 kHz')
```

```python
pi = np.pi
sin = np.sin
t = np.arange(0,1,1/fs)
y = 0.5*sin(2*f1*pi*t) + 0.4*sin(2*f2*pi*t) + 0.2*sin(2*f3*pi*t) + 0.3*sin(2*f4*pi*t) + 0.4*sin(2*f5*pi*t)
```

```python
fig1,ax = plt.subplots(figsize = [15,10])
ax.plot(t,y)
plt.show()
```

```python
Y,deltaf = getComponents(y,fs,t)
fig2,ax = plt.subplots(figsize = [15,10])
freqIndex = np.arange(0,len(t),1)
ax.plot(freqIndex,np.abs(Y))
plt.show()
```

```python
fs = 50.001e3
t = np.arange(0,1,1/fs)
y2 = 0.5*sin(2*f1*pi*t) + 0.4*sin(2*f2*pi*t) + 0.2*sin(2*f3*pi*t) + 0.3*sin(2*f4*pi*t) + 0.4*sin(2*f5*pi*t)
Y2,deltaf2 = getComponents(y2,fs,t)
fig2,ax = plt.subplots(figsize = [15,10])
freqIndex = np.arange(0,len(t),1)
ax.plot(freqIndex,np.abs(Y2))
plt.show()
```

```python
fs = 45.1e3
t = np.arange(0,1,1/fs)
y3 = 0.5*sin(2*f1*pi*t) + 0.4*sin(2*f2*pi*t) + 0.2*sin(2*f3*pi*t) + 0.3*sin(2*f4*pi*t) + 0.4*sin(2*f5*pi*t)
Y3,deltaf3 = getComponents(y3,fs,t)
fig2,ax = plt.subplots(figsize = [15,10])
freqIndex = np.arange(0,len(t),1)
ax.plot(freqIndex,np.abs(Y3))
plt.show()
```
